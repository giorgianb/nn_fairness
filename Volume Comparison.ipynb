```python
import json
from collections import defaultdict
import matplotlib.pyplot as plt
import numpy as np
```

```python
sd_results = defaultdict(list)
np_results = defaultdict(list)
```

```python
block_qhull = json.load(open('results/volume-compare/compas-block_qhull.json'))
for k, v in block_qhull.items():
    sd_results[k].append(v['Symmetric Difference'])
    np_results[k].append(v['Net Preference'])
```

```python
bounding_box = json.load(open('results/volume-compare/compas-true.json'))
for k, v in bounding_box.items():
    sd_results[k].append(v['Symmetric Difference'])
    np_results[k].append(v['Net Preference'])
```

```python
y0 = []
y1 = []
for v in sd_results.values():
    y0.append(v[0])
    y1.append(v[1])

plt.plot(y0)
plt.figure()
plt.plot(y1)
```

```python
y0 = []
y1 = []
for v in np_results.values():
    y0.append(v[0])
    y1.append(v[1])

plt.plot(y0)
plt.figure()
y1 = np.array(y1)
plt.plot(y1)
```
